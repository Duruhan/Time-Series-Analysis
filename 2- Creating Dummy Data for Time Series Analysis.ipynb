```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
idx = pd.date_range('1992-11-10', periods=150, freq='M')
y = np.ones(150)
df = pd.DataFrame(y, index=idx, columns=['Line'])

y = np.arange(1,151) 
df['Slope'] = y

smallNoice = []
for i in range(150):
    number = np.random.rand() * np.random.choice((1,-1))
    smallNoice.append(number)
df['SmallNoice'] = np.array(y + smallNoice)

largeNoice = []
for i in range(150):
    number = np.random.rand() * np.random.choice((1,-1)) * 10
    largeNoice.append(number)
df['LargeNoice'] = np.array(y + largeNoice)

seasonal = [0,1,2,1] * 40
seasonal = seasonal[:150]
df['Seasonal'] = seasonal

numbers = []
for i in range(150):
    number = np.random.rand() * np.random.choice((1,-1))
    numbers.append(number)
df['SeasonalHard'] = seasonal * (y + numbers)

numbers = []
for i in range(150):
    number = np.random.rand() * np.random.choice((1,-1))
    numbers.append(number)
df['SeasonalSoft'] = seasonal*np.ones(150)*30 + (y + numbers)

numbers = []
for i in range(150):
    number = np.random.rand() * np.random.choice((1,-1))
    numbers.append(number)
df['SeasonalRandom'] = seasonal*np.random.randn(150)*10 + (y + numbers)

a = np.random.normal(size = 150); a = np.cumsum(a); df['RandomA'] = a
b = np.random.normal(size = 150); b = np.cumsum(b); df['RandomB'] = b
c = np.random.normal(size = 150); c = np.cumsum(c); df['RandomC'] = c
```

```python
df.plot(figsize = (12,6));
```

```python
df.to_csv('Dummy Data for Time Series Analysis.csv', index=False)
```
